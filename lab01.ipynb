```python
import numpy as np
list = [9,8,4,9,5,7,1,3,2,7]
array1 = np.array(list)
array1
```

```python
#import numpy as np
list = [8,3,4,1,1,6,8,6,7,3]
arr = np.array(list)
arr
```

```python
print("arr mul array1 =", array1 * arr)
print("arr div array1 =" ,array1 / arr)
print("arr sub array1 =",array1 - arr)
print("arr add array1 =",array1 + arr)
```

```python
#  functions
print("sine function =",np.sin(array1))
print("log function =",np.log(array1))
print("base=10 log operation =",np.log10(array1))
print("base=2 log operation =",np.log2(array1))
# exop0nent
print("exp operation =",np.exp(array1))
```

```python
stu = ['sai','ram','akhil','nani']
age = [19,20,19,18]
weg = [90,89,78,67]
hig = [123,234,345,456]
import matplotlib.pyplot as plt
plt.scatter(weg,hig)
plt.title('releationship between weg & age of patients')
plt.ylabel('hig in cm')
plt.xlabel('weg of patients in kgs')

```

```python
plt.scatter(age,hig)
plt.title('releationship between age of patients')
plt.ylabel('hig in cm')
plt.xlabel('age of patients in kgs')
```

```python
plt.figure(figsize=(6,4))
plt.title("people weight in kgs",fontsize=16,fontstyle='italic')
# main plot bar
plt.bar(x=stu,height=weg,width=0.6,color='blue',edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("weight in kgs",fontsize=15)
```
